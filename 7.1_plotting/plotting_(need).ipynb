```python
from matplotlib import pyplot as plt
import numpy as np 
plt.style.use('default')
```

```python
time = np.arange(10)
y_2 = time**2
y_3 = time**3

plt.plot(time,y_2, label='$y^2$', linestyle="--", color= '#7fc97f')
plt.plot(time,y_3, label='$y^3$', color='blue')
plt.xlabel('time(s)')
plt.ylabel('test(units)')
plt.title("my first plot")
plt.grid(alpha=0.25)        #grid transparency 
plt.legend()                # draw grid 
plt.savefig('myfirstplot.png') #saves file under plotting need, can change the dpi 
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Let's start plotting
plt.plot(x, y_1, label='Y values',c='red', ls='dashed', lw = 1)
plt.plot(x, y_2, label='Y values',c="blue" )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error, label="Y3 with errors", color="green")

#labelling after 
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)


plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('https://berkeley-earth-temperature.s3.us-west-1.amazonaws.com/Global/Complete_TAVG_daily.txt',skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.8)
plt.ylim([-4, 4])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
#horizontal dashed line. 1st arguement 
```

## 4.2 Styles

```python
plt.style.use('fivethirtyeight')

data = np.loadtxt('https://berkeley-earth-temperature.s3.us-west-1.amazonaws.com/Global/Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-4, 4])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
