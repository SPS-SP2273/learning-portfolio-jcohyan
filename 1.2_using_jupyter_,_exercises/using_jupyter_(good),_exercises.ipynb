### Exercise 3 (Timing Stuff) The two code examples below achieve the same end result. ### 

```python
squares = [i**2 for i in range(10)]
print ([i**2 for i in range(10)])
```

```python
squares = list()
for i in range(10):
    squares.append(i**2)
print(squares)
```

Using %%timeit to time the execution of a cell: 

```python
%%timeit
squares = [i**2 for i in range(10)]
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
```

The result shows the time taken to run the cell 1 000 000 times in microseconds. It also shows the standard deviation of time taken. The results show that the first script takes slightly less time to execute but it has a higher standard deviation.
